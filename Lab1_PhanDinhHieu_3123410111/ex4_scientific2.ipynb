## Nội dung bài
Hiển thị 2 đồ thị sin(x), cos(x) trong khoảng $x \in \left[-\pi, \pi\right]$
## Bài làm 

### Import thư viện

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Sinh dữ liệu hàm số 

```python
X=np.linspace(-np.pi,np.pi,256,endpoint=True)
C,S=np.cos(X),np.sin(X)
```

```python
print(f'+X:|X|={len(X)},x in [{np.min(X):.2f},{np.max(X):.2f}],10 samples : {X[0:10]}')
print(f'+ Cos(X): |C|={len(C)}, C in [{np.min(C):.2f}, {np.max(C):.2f}], 10 samples: {C[0:10]}')
print(f'+ Sin(X): |S|={len(S)}, S in [{np.min(S):.2f}, {np.max(S):.2f}], 10 samples: {S[0:10]}')
```

### Hiển thị đồ thị

### Hiển thị đồ thị

##### Giới hạn của trục X

```python
plt.xlim(X.min() * 1.1, X.max() * 1.1);
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']);
```

##### Giới hạn của trục Y

```python
plt.ylim(C.min() * 1.1, C.max() * 1.1);
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$']);
plt.show()
```

##### Trực quan hóa dữ liệu

```python
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-");
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-");plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-");
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-");
```
